## PEI Election Data

Using the data from [Elections PEI - Open Data](https://www.electionspei.ca/resources/open-data), and more specifically [Yearly Political Party Contributions - Open Data](https://www.electionspei.ca/yearly-political-party-contributions-open-data), we create some visualizations of the data.

After cleaning the data, we can begin the visualizations, etc.

```python
# Dependencies.
import pandas as pd
```

```python
# Data.
df = pd.read_csv('../resources/transformed_data/election_contributions_transformed.csv')
df
```

```python
# Get only those in PEI.
pei_df = df.loc[df.Province == 'PE']
pei_df.head()
```

```python
# Get total contributions from each community in PEI - all years.
total_pei_df = pd.DataFrame(pei_df.groupby(['Community'])['Amount'].sum())
total_pei_df.to_csv('../resources/transformed_data/Election_Contribution_by_Community.csv')
total_pei_df.head()
```

```python
# Get total contributions from each community in PEI - all years by party.
party_pei_df = pd.DataFrame(pei_df.groupby(['Community', 'Party'])['Amount'].sum())
party_pei_df.to_csv('../resources/transformed_data/Election_Contribution_by_Community_by_Party.csv')
party_pei_df.head()
```

```python
# Get total amount from each community each year.
year_pei_df = pd.DataFrame(pei_df.groupby(['Year', 'Community', 'Party'])['Amount'].sum())
year_pei_df.to_csv('../resources/transformed_data/Election_Contribution_by_Year_by_Community_by_Party.csv')
year_pei_df
```
